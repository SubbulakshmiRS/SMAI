```python
import csv
import numpy as np 
import collections
from sklearn.metrics import accuracy_score, mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt
```

```python
class Airfoil(object):
    def __init__(self, itNum, alpha, coeff):
        self.itNum = itNum
        self.alpha = alpha
        self.coeff = coeff

    def gradient_descent(self, trainData):
        x = trainData[:,0:6]
        y = trainData[:,6]
        m, n = trainData.shape
        cost_list = []
        for i in range(0, self.itNum):
            prediction = np.dot(x, self.coeff)
            error = prediction - y
            cost = 1/(2*m) * np.dot(error.T, error)
            cost_list.append(cost)
            self.coeff = self.coeff - (self.alpha * (1/m) * np.dot(x.T, error)) 
        return cost_list

    def prepData(self, trainData):
        r,c = trainData.shape
        x = trainData[:,0:5]
        y = trainData[:,5]
        x_normed = (x - x.min(0)) / x.ptp(0)
        trainData = np.c_[ x_normed, y ] 
        trainData = np.c_[ np.ones(r), trainData ] 
        return trainData

    def validate(self, validateData):
        x = validateData[:,0:6]
        y = validateData[:,6]
        predictions = np.dot(x, self.coeff)
        self.stats(y, predictions)

    def stats(self, test_labels, predictions ):
        print ("Mean squared error", mean_squared_error(test_labels, predictions))
        # print ("Accuracy score", accuracy_score(test_labels, predictions))
        print ("Mean absolute error", mean_absolute_error(test_labels, predictions))

    def predict(self, TestFile):
        # print("predict")
        with open(TestFile,'r') as fT:
            readerT  = csv.reader(fT, quoting=csv.QUOTE_NONNUMERIC)
            TestData = np.array(list(readerT))
            TestData = self.prepData(TestData)

    def train(self, DataFile):
        with open(DataFile,'r') as fD:
            readerD = csv.reader(fD, quoting=csv.QUOTE_NONNUMERIC)
            trainData = np.array(list(readerD))
            np.random.shuffle(trainData)
            trainData = self.prepData(trainData)
            print("Learning rate ",self.alpha)
            print("Coeffients init", self.coeff)
            validateData = trainData[0:200]
            trainData = trainData[200:]
            cost_list = self.gradient_descent(trainData)
            print("Number of iterations ", self.itNum)
            self.validate(validateData)
            return cost_list
```

```python
itNum = 500000
alpha1 = 0.00001
alpha2 = 0.0001
alpha3 = 0.001
coeff_zeros = np.zeros(6)
coeff_ones = np.ones(6)
np.random.seed(10)
coeff_random = np.random.rand(6)
```

Plots between the costs for each iterations

```python
fig,ax =  plt.subplots(figsize=(10, 10))
# %matplotlib inline
x = np.arange(1,itNum+1)
model = Airfoil(itNum, alpha1, coeff_random)
costListR = model.train('./Question-3/airfoil.csv')
ax.plot(x,costListR, 'C1', label='C1')
model = Airfoil(itNum, alpha1, coeff_zeros)
costListZ = model.train('./Question-3/airfoil.csv')
ax.plot(x,costListZ, 'C2', label='C2')
model = Airfoil(itNum, alpha1, coeff_ones)
costListO = model.train('./Question-3/airfoil.csv')
ax.plot(x,costListO, 'C3', label='C3')
ax.set(xlabel='iteration', ylabel='Cost',
       title='Across different initial coeffient values')
plt.show()
```

When you want to apply regression on a data containing a feature with missing data such that:
    (a) When 0-0.5% of values are missing of that feature
        The solution is to remove those samples having the missing feature value
    (b) When 8-10% of values are missing of that feature
        The solution is to replace the missing values with the mean/median/mode of that feature column.
    (c) When 60-70% of values are missing of that feature
        The solution is to remove the feature column from the database itself.
